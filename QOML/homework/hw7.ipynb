```python
import dimod
from dwave.samplers import SimulatedAnnealingSampler

%matplotlib inline
```

# Problem 1

```python
h = {0: 0, 1: 0, 2: 0, 3: 0}

J = {
    (0, 1): 1,
    (1, 2): 1,
    (2, 3): 1,
    (3, 0): 1
}

sampler = SimulatedAnnealingSampler()
sampleset = sampler.sample_ising(h, J)
print("Best sample:", sampleset.first.sample)
print("Best energy:", sampleset.first.energy)

# print num_reads and time results
for r in [10, 100, 500, 1000]:
    for t in [5, 20, 50, 100]:
        sampleset = sampler.sample_ising(h, J, num_reads=r, annealing_time=t)
        print(r, t, sampleset.first.sample, sampleset.first.energy)
```

# Problem 2

```python
Q = {
    (0, 0): -2,
    (1, 1): -2,
    (2, 2): -2,
    (3, 3): -2,
    (0, 1):  2,
    (1, 2):  2,
    (2, 3):  2,
    (0, 3):  2
}

sampleset = sampler.sample_qubo(Q)

print("Best sample:", sampleset.first.sample)
print("Best energy:", sampleset.first.energy)

# print num_reads and time results
for r in [10, 100, 500, 1000]:
    for t in [5, 20, 50, 100]:
        sampleset = sampler.sample_qubo(Q, num_reads=r, annealing_time=t)
        print(r, t, sampleset.first.sample, sampleset.first.energy)
```

# Problem 3

```python
linear = {
    'x0': -8,
    'x1': -11,
    'x2': -6,
    'x3': -4,
    'y0': 0,
    'y1': 0,
    'y2': 0,
    'y3': 0
}

weights = {
    'x0': 4,
    'x1': 5,
    'x2': 3,
    'x3': 2,
    'y0': 1,
    'y1': 2,
    'y2': 4,
    'y3': 8
}

# build penalty term 
bqm = dimod.BinaryQuadraticModel({}, {}, 0.0, dimod.BINARY)

for v in linear:
    bqm.add_variable(v, linear[v])

# expand penalty
for u in weights:
    # diagonal term
    bqm.add_variable(u, 30 * (weights[u]**2 - 20*weights[u]))

for u in weights:
    for v in weights:
        if u < v:
            bqm.add_interaction(u, v, 2 * 30 * weights[u] * weights[v])

# constant
bqm.offset += 100 * 30

# solve 
sampleset = dimod.ExactSolver().sample(bqm)

print("Best sample:", sampleset.first.sample)
print("Best energy:", sampleset.first.energy)

# print num_reads and time results
for r in [10, 100, 500, 1000]:
    for t in [5, 20, 50, 100]:
        sampleset = dimod.ExactSolver().sample(bqm, num_reads=r, annealing_time=t)
        print(r, t, sampleset.first.sample, sampleset.first.energy)
```

```python
Q = {}
c = {
    'x0': 4, 'x1': 5, 'x2': 3, 'x3': 2,
    'y0': 1, 'y1': 2, 'y2': 4, 'y3': 8
}

value = {'x0': 8, 'x1': 11, 'x2': 6, 'x3': 4}

vars_list = list(c.keys())

# linear terms
for v in vars_list:
    Q[(v, v)] = 30 * (c[v]**2 - 20*c[v]) - value.get(v, 0)

# quadratic terms
for i in range(len(vars_list)):
    for j in range(i + 1, len(vars_list)):
        vi, vj = vars_list[i], vars_list[j]
        Q[(vi, vj)] = 2 * 30 * c[vi] * c[vj]

bqm = dimod.BinaryQuadraticModel.from_qubo(Q, offset=3000)

sampleset = sampler.sample(bqm, num_reads=1000)

best = sampleset.first.sample
energy = sampleset.first.energy

print("Best sample:", best)
print("Energy:", energy)

x = [best['x0'], best['x1'], best['x2'], best['x3']]
weight = 4*x[0] + 5*x[1] + 3*x[2] + 2*x[3]
val = 8*x[0] + 11*x[1] + 6*x[2] + 4*x[3]

print("Total weight:", weight)
print("Total value:", val)
```
